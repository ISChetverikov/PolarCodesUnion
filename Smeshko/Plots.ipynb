```python
import matplotlib.pyplot as plt
%matplotlib inline

import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('numbers.csv', index_col=0)
df.columns = ['1', '2', '4', '8', '16', '32', '64', '128', '256', '512']

dfx = pd.read_excel('Book1.xlsx', index_col=0)
# dfx = dfx.iloc[:15, 11:]
```
